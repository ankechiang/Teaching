# COMPSCI753: Vectorize New Articles into Bitvectors

```python
f = open("./bbc/text_all.csv", "r")
train_documents = f.readlines()
f = open("./bbc/text_query.csv", "r")
test_documents = f.readlines()
```

```python
print(len(train_documents), len(test_documents))
```

```python
train_documents[1]
```

```python
document_id = []
sample_documents = []
label = []
for document in train_documents:
    line = document.strip().split("\t")
    if line[1] == 'Text': continue
    document_id.append(line[0])
    sample_documents.append(line[1])
    label.append(line[-1])
for document in test_documents:
    line = document.strip().split("\t")
    if line[1] == 'Text': continue
    document_id.append(line[0])
    sample_documents.append(line[1])
    label.append(line[-1])
```

```python
print(len(document_id), len(sample_documents), len(label))
```

## Bag of Words
### scikit-learn library provides CountVectorizer class to perform this vectorization process. Firstly, Identify Unique words...

```python
from sklearn.feature_extraction.text import CountVectorizer

# Identify Unique words
vectorizer = CountVectorizer()
vectorizer.fit(sample_documents)
print(f'vector vocabulary - {vectorizer.vocabulary_}\n')
```

### Perform Vectorization: find the frequency count of each unique word, and zero otherwise. Each article becomes a vector of fixed size.


```python
# encode document
vector = vectorizer.transform(sample_documents)
# summarize encoded vector
print(f'vector shape: {vector.shape}\n')
print(f'article vector\n {vector.toarray()}')
```

```python
import csv

f = open('./bbc/bitvector_all.csv', 'w')
for i, arr in enumerate(vector.toarray()):

    feature = []
    for a in arr:
        if a>1: feature.append('1')
        elif a==1: feature.append('1')
        else: feature.append('0')


    f.write('{0}\t{1}\t{2}\n'.format(i+1, '\t'.join(feature), label[i]))
    
f.close()    
    
```
